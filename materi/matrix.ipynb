# Matrix

## Definisi matrix

Matriks adalah susunan angka yang disusun dalam bentuk baris dan kolom, seperti tabel. Setiap angka dalam matriks disebut elemen.
Misalnya, jika kita punya matriks:

$$
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$

Angka 1, 2, 3, dan 4 adalah elemen matriks.
Matriks ini memiliki 2 baris (horizontal) dan 2 kolom (vertikal).


## Beberapa Jenis Matrix Penting

### 1.matrix persegi

Matriks Persegi adalah jenis matriks di mana jumlah barisnya sama dengan jumlah kolomnya.

$$
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

### 2.matrix baris

Matriks Baris (Row Matrix) adalah jenis matriks yang memiliki hanya satu baris.

$$
\begin{bmatrix}
a & b & c & d
\end{bmatrix}
$$

### 3.matrix kolom

Matriks Kolom (Column Matrix) adalah jenis matriks yang memiliki hanya satu kolom.

$$
\begin{bmatrix}
a \\ b \\ c \\ d
\end{bmatrix}
$$

### 4.matrix nol

Matriks Nol (Zero Matrix) adalah jenis matriks di mana semua elemen dalam matriks tersebut bernilai nol.

$$
\begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}
$$

### 5.matrix identitas

Matriks Identitas (Identity Matrix) adalah jenis matriks persegi khusus yang memiliki elemen diagonal utama bernilai 1, dan semua elemen lainnya bernilai 0.

$$
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
$$

### 6.matrix diagonal

Matriks Diagonal (Diagonal Matrix) adalah jenis matriks persegi yang memiliki elemen-elemen non-nol hanya pada diagonal utama (diagonal dari kiri atas ke kanan bawah), sementara elemen-elemen lainnya adalah nol.

$$
\begin{bmatrix}
a & 0 & 0 \\
0 & b & 0 \\
0 & 0 & c
\end{bmatrix}
$$

### 7.matrix skalar

Matriks Skalar (Scalar Matrix) adalah jenis matriks diagonal khusus di mana semua elemen diagonalnya memiliki nilai yang sama.

$$
\begin{bmatrix}
a & 0 & 0 \\
0 & a & 0 \\
0 & 0 & a
\end{bmatrix}
$$

## Operasi aritmatika matrix
### 1. Penjumlahan
* Konsep Penjumlahan Matrix

Dua matriks bisa dijumlahkan jika memiliki ukuran yang sama (jumlah baris dan kolom sama).Penjumlahan dilakukan dengan menambahkan elemen-elemen yang bersesuaian.

**Contoh:**

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
,B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}\\
A + B =
\begin{bmatrix}
1 + 5 & 2 + 6 \\
3 + 7 & 4 + 8
\end{bmatrix}
=
\begin{bmatrix}
6 & 8 \\
10 & 12
\end{bmatrix}
$$

**KEDUA MATRIX HARUS SAMA**

```python
def penjumlahan(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(A[0])):
            baris.append(A[i][j] + B[i][j])
        hasil.append(baris)
    return hasil

# Input ukuran matriks
baris = int(input("Masukkan jumlah baris: "))
kolom = int(input("Masukkan jumlah kolom: "))

# Input matriks A
print("Masukkan elemen matriks A:")
A = []
for i in range(baris):
    baris_A = []
    for j in range(kolom):
        baris_A.append(int(input(f"A: ")))
    A.append(baris_A)

# Input matriks B
print("Masukkan elemen matriks B:")
B = []
for i in range(baris):
    baris_B = []
    for j in range(kolom):
        baris_B.append(int(input(f"B: ")))
    B.append(baris_B)

# Hitung hasil penjumlahan
proses = penjumlahan(A, B)

# Cetak hasil
print("Hasil penjumlahan matriks:")
for baris in proses:
    print(baris)
```

### 2. Pengurangan
* Konsep Pengurangan matrix

Sama seperti penjumlahan, tetapi setiap elemen dikurangi dengan elemen yang bersesuaian (jumlah baris dan kolom sama).pengurangan dilakukan dengan mengurangi elemen-elemen yang bersesuaian.

**Contoh:**

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
,B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}\\
A - B =
\begin{bmatrix}
1 - 5 & 2 - 6 \\
3 - 7 & 4 - 8
\end{bmatrix}
=
\begin{bmatrix}
-4 & -4 \\
-4 & -4
\end{bmatrix}
$$

```python
def pengurangan(A, B):
    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(A[0])):
            baris.append(A[i][j] - B[i][j])
        hasil.append(baris)
    return hasil

# Input ukuran matriks
baris = int(input("Masukkan jumlah baris: "))
kolom = int(input("Masukkan jumlah kolom: "))

# Input matriks A
print("Masukkan elemen matriks A:")
A = []
for i in range(baris):
    baris_A = []
    for j in range(kolom):
        nilai = int(input(f"A: "))
        baris_A.append(nilai)
    A.append(baris_A)

# Input matriks B
print("Masukkan elemen matriks B:")
B = []
for i in range(baris):
    baris_B = []
    for j in range(kolom):
        nilai = int(input(f"B: "))
        baris_B.append(nilai)
    B.append(baris_B)

# Hitung hasil pengurangan
proses = pengurangan(A, B)

# Cetak hasil
print("Hasil pengurangan matriks:")
for i in range(len(proses)):
    print(proses[i])
```

### 3. Perkalian
* Konsep Pekalian dua matrix

Untuk mengalikan dua matriks A × B, jumlah kolom matriks pertama harus sama dengan jumlah baris matriks kedua. Jika $A$ berukuran $m$ × $n$ dan $B$ berukuran $m$ × $p$

**Contoh:**

$$
A =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
,B =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}\\
AB =
\begin{bmatrix}
(1 × 5 + 2 × 7) & (1 \times 6 + 2 × 8) \\
(3 × 5 + 4 × 7) & (3 \times 6 + 4 × 8)
\end{bmatrix} \\
=
\begin{bmatrix}
19 & 22 \\
43 & 50
\end{bmatrix}
$$

```python
def perkalian(A, B):
    if len(A[0]) != len(B):
        raise ValueError("Jumlah kolom matriks pertama harus sama dengan jumlah baris matriks kedua!")

    hasil = []
    for i in range(len(A)):
        baris = []
        for j in range(len(B[0])):
            total = 0
            for k in range(len(B)):
                total += A[i][k] * B[k][j]
            baris.append(total)
        hasil.append(baris)
    return hasil

# Input ukuran matriks A
baris_A = int(input("Masukkan jumlah baris matriks A: "))
kolom_A = int(input("Masukkan jumlah kolom matriks A: "))

# Input matriks A
print("Masukkan elemen matriks A:")
A = []
for i in range(baris_A):
    baris = []
    for j in range(kolom_A):
        nilai = int(input(f"A: "))
        baris.append(nilai)
    A.append(baris)

# Input ukuran matriks B
baris_B = kolom_A  # Agar perkalian valid, jumlah baris B harus sama dengan jumlah kolom A
kolom_B = int(input("Masukkan jumlah kolom matriks B: "))

# Input matriks B
print("Masukkan elemen matriks B:")
B = []
for i in range(baris_B):
    baris = []
    for j in range(kolom_B):
        nilai = int(input(f"B: "))
        baris.append(nilai)
    B.append(baris)

# Hitung hasil perkalian
proses = perkalian(A, B)

# Cetak hasil
print("Hasil perkalian matriks:")
for i in range(len(proses)):
    print(proses[i])
```

* konsep Perkalian matrix dengan skalar

Setiap elemen dalam matriks dikalikan dengan suatu bilangan (skalar).

**Contoh:**

$$
3A =
\begin{bmatrix}
3 × 1 & 3 × 2 \\
3 × 3 & 3 × 4
\end{bmatrix}
=
\begin{bmatrix}
3 & 6 \\
9 & 12
\end{bmatrix}
$$

```python
def perkalian_skalar(matriks, skalar):
    hasil = []
    for i in range(len(matriks)):
        baris = []
        for j in range(len(matriks[0])):
            baris.append(matriks[i][j] * skalar)
        hasil.append(baris)
    return hasil

# Input ukuran matriks
baris = int(input("Masukkan jumlah baris: "))
kolom = int(input("Masukkan jumlah kolom: "))

# Input matriks
print("Masukkan elemen matriks:")
matriks = []
for i in range(baris):
    baris_matriks = []
    for j in range(kolom):
        nilai = int(input(f"Matriks[{i}][{j}]: "))
        baris_matriks.append(nilai)
    matriks.append(baris_matriks)

# Input skalar
skalar = int(input("Masukkan nilai skalar: "))

# Hitung hasil perkalian skalar
proses = perkalian_skalar(matriks, skalar)

# Cetak hasil
print("Hasil perkalian skalar:")
for i in range(len(proses)):
    print(proses[i])
```

## determinan Matrik
Determinan adalah suatu nilai skalar yang dapat dihitung dari elemen-elemen suatu matriks persegi (matriks dengan jumlah baris dan kolom yang sama). Determinan sering digunakan dalam berbagai aplikasi, seperti menemukan invers matriks, menyelesaikan sistem persamaan linear, dan analisis transformasi linier.

Contoh Determinan Matriks Ordo 2×2

Contoh matrix :

$$
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$

Rumus determinan:

$$
\begin{array}{cc}
det(A) = (a \times d) - (b \times c) = hasil
\end{array}
$$

Sifat-Sifat Determinan Matriks

1. Determinannya 1 untuk Matriks Identitas

2. Jika Ada Baris f Kolom Nol, Determinan=0

3. Jika Dua Baris/Kolom Sama atau Proporsional, Determinan=0

4. Jika Baris f Kolom Dikalikan,Determinan Juga Dikalikan

5. Tukar Dua BarisfKolom, Determinan Berubah Tanda

6. Matriks Segitiga: Determinan=Perkalian Diagonal
### minor matrik
Minor dari suatu elemen dalam matriks adalah determinan dari submatriks yang diperoleh dengan menghapus baris dan kolom yang memuat elemen tersebut.
Contoh Minor Matriks

matrix:

$$
A =
\begin{pmatrix}
1 & 2 & 3 \\
4 & 5 & 6\\
7 & 8 & 9
\end{pmatrix}
$$

Minor dari elemen $a_{11}$ = 1 adalah determinan dari submatriks yang diperoleh dengan menghapus baris pertama dan kolom pertama:

$$
M_{11} =
\begin{pmatrix}
5 & 6\\
8 & 9
\end{pmatrix} = (5 \times 9) - (6 \times 8) = 45 -48 = -2
$$

### cofaktor Matrik
Kofaktor dari elemen ( $a_{ij}$ ) dalam suatu matriks adalah minor dari elemen tersebut yang dikalikan dengan faktor tanda (-1) berpangkat ( i + j ):

$$
\begin{array}{cc}
C_{ij} = (-1)^{i+j}\ M_{ij}
\end{array} 
$$

**Contoh kofaktor**

Dari contoh sebelumnya, kofaktor untuk elemen $(a_{12}=2)$ adalah:

$$
M_{12} =
\begin{bmatrix}
4 & 6\\
7 & 9
\end{bmatrix}
$$

Determinannya dihitung sebagai:

$$
\begin{array}{cc}
det(M_{12}) = (4 \times 9) - (6 \times 7) = 36 - 42 = -6
\end{array} 
$$

$$
\begin{array}{cc}
C_{12} = (-1)^{1+2} \times - (-6) = (-1)_3 \times -(-6) = 6
\end{array} 
$$

Minor dan kofaktor memiliki peran penting dalam menghitung determinan matriks, terutama untuk matriks berordo lebih dari 2 x 2

### mencari determinan dengan konsep minor dan cofactor matrik
**Minor**

Minor dari elemen $𝑎_{𝑖𝑗}$  adalah determinan dari matriks 2x2 yang diperoleh dengan menghapus baris ke- $𝑖$ dan kolom ke- $𝑗$ dari matriks asal.

**kofaktro**

Cofactor dari elemen $𝑎_{ij}  adalah:

$$
\begin{array}{cc}
C_{ij} = (-1)^{i+j}\ \cdot M_{ij}
\end{array} 
$$

di mana $M_{ij}$  adalah minor dari elemen tersebut.

Untuk menghitung determinan menggunakan konsep minor dan kofaktor, kita menggunakan ekspansi kofaktor, yaitu:

$$
\begin{array}{cc}
det(A) = a_{i1}C_{i1}+a_{i2}C_{i2} + ... + a_{in}C_{in}
\end{array}
$$

**contoh:**

1. Determinan matrix 3 \times 3:

$$A =
\begin{bmatrix}
1 & 2 & 3\\
0 & 4 & 5\\
1 & 0 & 6
\end{bmatrix}
$$

Elemen $a_{11}$ = 1

Minor $M_{11}$:

$$
\begin{vmatrix}
4 & 5\\
0 & 6
\end{vmatrix}
= (4)(6)-(0)(5)=24
$$

Cofaktor $C_{11}$ = (+1) \cdot 24 = 24

Elemen $a_{12}$ = 2

Minor $M_{12}$:

$$
\begin{vmatrix}
0 & 5\\
1 & 6
\end{vmatrix}
= (0)(6)-(1)(5)=-5
$$

Cofaktor $C_{12}$ (-1) \cdot (-5) = 5

Elemen $a_{13}$ = 3

Minor $M_{13}$:

$$
\begin{vmatrix}
0 & 4\\
1 & 0
\end{vmatrix}
= (0)(0)-(1)(4)=-4
$$

Cofaktor $C_{13}$ (+1) \cdot (-4) = 4

Determinan matrix 3x3

$$
\begin{array}{cc}
det(A) = a_{11}C_{11} + a_{12}C_{12} + a_{13}C_{13}\\
= (1)(24) + (2)(5) + (3)(-4) = 24 + 10 - 12 = 22
\end{array}
$$

2. Determinan matrix 4 \times 4:

$$A =
\begin{bmatrix}
1&2&0&3\\
4&0&1&2\\
0&1&2&1\\
3&0&1&1
\end{bmatrix}
$$

Gunakan ekspansi baris pertama:

$$
\begin{array}{cc}
det(B) = 1C_{11}-2C_{12}+3C_{13}-4C_{14}
\end{array}
$$

Hitung masing-masing **minor** dan **kofaktor**, lalu gunakan determinan 3 \times 3 seperti pada contoh sebelumnya.
